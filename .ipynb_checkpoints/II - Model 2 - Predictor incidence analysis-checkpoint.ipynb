```python
import numpy as np
import pandas as pd
```

```python
donors = pd.read_csv('data/donations_eng.csv')
```

```python
donors.head()
```

# Discretising variables for graphs

```python
def check_discretise(table, variable, threshold):
    return len(table.groupby(variable)) >= threshold

print(check_discretise(donors, 'age', 5))
print(check_discretise(donors, 'country_USA', 5))
```

```python
threshold = 5
number_bins = 5
disc_variables = []

for v in donors.columns:
    if check_discretise(donors, v, threshold):
        new_variable = 'disc_' + v
        donors[new_variable] = pd.qcut(donors[v], number_bins)
        disc_variables.append(v)
```

```python
disc_variables
```

```python
donors['country'] = donors[['country_USA', 'country_India', 'country_UK']].idxmax(1)
```

```python
def check_medium_income(row):
    if row['income_high']==0 and row['income_low']==0:
        return 1
    else:
        return 0
    
donors['income_med'] = donors.apply(check_medium_income, axis=1)
print(sum(donors.income_high)+sum(donors.income_low)+sum(donors.income_med))
# 24993 is correct
```

```python
donors['income'] = donors[['income_low', 'income_med', 'income_high']].idxmax(1)
```

```python
donors.head()
```

```python
features_disc = list(set(donors.columns) - set(disc_variables) - set(['country_India', 'country_USA', 'country_UK'])
                    - set(['income_low', 'income_med', 'income_high']))
features_disc.remove('target')
print(features_disc)
```

# asds

```python
def create_pig_table(df, target, variable):
    groups = df[[target, variable]].groupby(variable)
    pig_table = groups[target].agg(Incidence='mean', Size='count').reset_index()
    return pig_table
```

```python
print(create_pig_table(donors, 'target', 'income'))
```

```python
print(create_pig_table(donors, 'target', 'disc_min_gift'))
```

```python
print(create_pig_table(donors, 'target', 'country'))
```

```python
pig_tables = {}
for f in features_disc:
    pig_tables[f] = create_pig_table(donors, 'target', f)
```

```python
import matplotlib.pyplot as plt
```

```python
pig_table = pig_tables['country']
#plt.rcParams['figure.figsize']=(5,5) 

# Y axis
pig_table['Size'].plot(kind='bar', width=0.5, color='lightgrey', edgecolor='none')
plt.ylabel('Size', rotation=0, rotation_mode='anchor', ha='right')

# X axis
plt.xticks(range(len(pig_table)), pig_table['country'])
plt.xlim(-0.5, len(pig_table)-0.5)
plt.xlabel('Country')

# Y axis seconday
pig_table['Incidence'].plot(secondary_y=True)
plt.ylabel('Incidence', rotation=0, rotation_mode='anchor', ha='left')

plt.show()
```

```python
for f in pig_tables.keys():
    pig_table = pig_tables[f]
    pig_table['Size'].plot(kind='bar', width=0.5, color='lightgrey', edgecolor='none')
    plt.ylabel('Size', rotation=0, rotation_mode='anchor', ha='right')
    plt.xticks(range(len(pig_table)), pig_table[f])
    plt.xlim(-0.5, len(pig_table)-0.5)
    plt.xlabel(f)
    pig_table['Incidence'].plot(secondary_y=True)
    plt.ylabel('Incidence', rotation=0, rotation_mode='anchor', ha='left')
    plt.show()
```
